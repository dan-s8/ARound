# I. New dataset with the demographic data by City

## 1. Importing the new dataset with the demographic data by cities

```python
df1 = pd.read_csv('/Users/jefferysun/Desktop/A_Round_RnA/US_DEMOGRAPIC_City/ACSDP5Y2022.DP05-Data.csv')
df1.columns
```

```python
df1.head()
```

## 2. Data Preprocessing

```python
df1[['Place_Name', 'State']] = df1['NAME'].str.split(',', n=1, expand=True)

# Cleaning up the name of the columns
for col in ['State', 'Place_Name']:
    df1[col] = df1[col].str.strip()  # Step 1: Remove leading and trailing spaces
    df1[col] = df1[col].str.replace('[^a-zA-Z0-9_]', '')  # Step 2: Remove special characters
    #df1[col] = df1[col].str.title()  # Step 3: capitalizes each word's first letter

# Choosing target columns
filtered_df = df1[['State', 'Place_Name', 'DP05_0001E', 'DP05_0044E']]
filtered_df.columns = ['State', 'City', 'Total_Population_City',  'Total_Asian_Population_City']
filtered_df = filtered_df.drop(filtered_df.index[0])
filtered_df.head(5)
```

## 3. Feature Creation

```python
filtered_df['Asian_Percentage_City'] = filtered_df['Total_Asian_Population_City'].astype(int) / filtered_df['Total_Population_City'].astype(int) * 100
filtered_df.head()
```

```python
# Preprocesing before Merging
filtered_df2 = filtered_df[filtered_df['State'].isin(state_list)].reset_index(drop=True)
filtered_df2['City'] = filtered_df2['City'].str.replace('city', '').str.replace('CDP', '').str.replace('borough', '').str.replace('town', '').str.replace('village', '').str.strip()
filtered_df2.reset_index(drop=True)
filtered_df2[filtered_df2['City']=='Kansas']
filtered_df2.loc[filtered_df2['City'] == 'Kansas', 'City'] = 'Kansas City'
filtered_df2.at[8000,'City']='Louisville'
filtered_df2.at[19406,'City']='Nashville'
filtered_df2.at[10529,'City']='Saint Paul'


# Inserting additional data manually 
new_row = pd.DataFrame([{'State': 'New York', 'City': 'Belmont Park', 'Total_Population_City': 8622467, 'Total_Asian_Population_City': 1247994, 'Asian_Percentage_City': 14.473746}, # New York
                        {'State': 'Montana', 'City': 'Bonner', 'Total_Population_City': 74627, 'Total_Asian_Population_City': 1214, 'Asian_Percentage_City': 1.626757},
                        {'State': 'New York', 'City': 'Brooklyn', 'Total_Population_City': 8622467, 'Total_Asian_Population_City': 1247994, 'Asian_Percentage_City': 14.473746}, # New York
                        {'State': 'Connecticut', 'City': 'Uncasville', 'Total_Population_City': 27596, 'Total_Asian_Population_City': 648, 'Asian_Percentage_City': 2.348166}, # New London
                        {'State': 'California', 'City': 'Hollywood', 'Total_Population_City': 35358, 'Total_Asian_Population_City': 1647, 'Asian_Percentage_City': 4.658069}, # West Hollywood
                        ])

filtered_df2 = pd.concat([filtered_df2, new_row], ignore_index=True)
```

## 4. Data Merging

```python
# Merging
merged_df = pd.merge(df, filtered_df2, on=['State', 'City'], how='left')

# The cities and states assoicated with null values in the new columns
merged_df[merged_df['Asian_Percentage_City'].isnull()].groupby(['City','State'])['Unnamed: 0'].count()
```

```python
# The number of rows including NULL value
merged_df[merged_df['Asian_Percentage_City'].isnull()].groupby(['City','State'])['Unnamed: 0'].count().sum()
```

## 5. Exporting merged dataset

```python
merged_df.to_csv('PlusSupplementaryFeatures_CityLevel.csv', index=False)
```

# II. New dataset with the demographic data by County

## 1. Importing Data

```python
df_county = pd.read_csv('/Users/jefferysun/Desktop/A_Round_RnA/US_DEMOGRAPIC_COUNTY/DECENNIALDHC2020.P8-Data.csv')
city_and_county = pd.read_csv('/Users/jefferysun/Desktop/A_Round_RnA/uscities_info/uscities.csv')
df_county.head()
```

```python
city_and_county.head(3)
```

## 2. Data Preprocessing


```python
# 1. County data
df_county[['County', 'State']] = df_county['NAME'].str.split(',', n=1, expand=True)

# Cleaning up the name of the columns
for col in ['State', 'County']:
    df_county[col] = df_county[col].str.strip()  # Step 1: Remove leading and trailing spaces
    df_county[col] = df_county[col].str.replace('[^a-zA-Z0-9_ ]', '')  # Step 2: Remove special characters
    #df_county[col] = df_county[col].str.title()  # Step 3: capitalizes each word's first letter

df_county['County'] = df_county['County'].str.replace(r'\bCounty of\b|\bThe County of\b|\bCounty\b', '', case=False, regex=True).str.strip()

# Choosing target columns
df_county = df_county[['State', 'County', 'P8_001N', 'P8_006N']]
df_county.columns = [ 'State', 'County', 'Total_Population_County', 'Total_Asian_Population_County']
df_county = df_county.drop(df_county.index[0])
df_county.head(5)
```

```python
city_and_county = city_and_county[['city', 'state_name', 'county_name']]
city_and_county = city_and_county.rename(columns={"city": "City", "state_name": "State", "county_name": "County"})
new_row = pd.DataFrame([{'City': 'Hollywood', 'State': 'California', 'County': 'Los Angeles'},
                        {'City': 'Belmont Park', 'State': 'New York', 'County': 'Nassau'},
                        {'City': 'Bonner', 'State': 'Montana', 'County': 'Missoula'},
                        {'City': 'Huntington', 'State': 'New York', 'County': 'Suffolk'},
                        {'City': 'Uncasville', 'State': 'Connecticut', 'County': 'New London'},
                        {'City': 'Saint Paul', 'State': 'Minnesota', 'County': 'Ramsey'}
                        ])
city_and_county = pd.concat([city_and_county, new_row], ignore_index=True)
city_and_county
```

```python
df_county['Asian_Percentage_County'] = df_county['Total_Asian_Population_County'].astype(int) / df_county['Total_Population_County'].astype(int) * 100
```

```python
# 2. City and County Data
city_and_county.head(3)
```

```python
# Merging
merged_df_county = pd.merge(df, city_and_county, on=['State', 'City'], how='left')
```

```python
merged_df_county.head(3)
```

```python
df_county_dem = pd.merge(merged_df_county, df_county, on=['State', 'County'], how='left')
df_county_dem.head(3)
```

```python
# Missing Value
df_county_dem[df_county_dem[['Total_Population_County','Asian_Percentage_County','Total_Asian_Population_County']].isnull()].count().any()
```
